## Logistic Regression

## Read the dataset

```python
path = (
    r"https://raw.githubusercontent.com/Sindhura-tr/Datasets/refs/heads/main/Iris.csv"
)
import pandas as pd

df = pd.read_csv(path)
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

## Separate X and Y features
Y: Species

```python
X = df.drop(columns=["Id", "Species"])
Y = df[["Species"]]
```

```python
X.head()
```

```python
Y.head()
```

## Data Preprocessing and Data Cleaning

```python
# Data Preprocessing and Data Cleaning
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = make_pipeline(SimpleImputer(strategy="mean"), StandardScaler()).set_output(
    transform="pandas"
)
```

```python
num_pipe
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre
```

## Split the data into training and testing

```python
# Train Test Split
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(
    X_pre, Y, train_size=0.85, random_state=42
)
```

```python
xtrain.head()
```

```python
xtest.head()
```

```python
ytrain.head()
```

```python
ytest.head()
```

## Model Building

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(xtrain, ytrain)
```

```python
model.classes_
```

```python
model.coef_
```

```python
model.intercept_
```

## Model Evaluation

```python
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

```python
ypreds = model.predict(xtest)
ypreds
```

```python
yprobs = model.predict_proba(xtest)
yprobs[:5]
```

## COnfusion Matrix, Classification Report

```python
from sklearn.metrics import ConfusionMatrixDisplay, classification_report

ConfusionMatrixDisplay.from_estimator(model, xtest, ytest)
```

```python
print(classification_report(ytest, ypreds))
```

```python
from sklearn.metrics import f1_score
```

```python
f1_test = f1_score(ytest, ypreds, average="macro")
f1_test
```

## Check for cross validated scores

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, xtrain, ytrain, cv=3, scoring="f1_macro")
scores
```

```python
scores.mean()
```
